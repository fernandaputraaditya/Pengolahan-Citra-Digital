```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
```

1. Pengambilan Citra

```python
img = cv2.imread('./bunga.jpg')
```

2. RGB

```python
# convert image to RGB color for matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show image with matplotlib
plt.imshow(img)
```

3. Grayscale

```python
# convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# grayscale image represented as a 2-d array
print(gray_img)
```

```python
# have to convert grayscale back to RGB for plt.imshow()
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

4. Cropping

```python
# crop image
img_cropped = gray_img[50:1300, 150:1400]
plt.imshow(cv2.cvtColor(img_cropped, cv2.COLOR_GRAY2RGB))
```

5. Histogram

```python
hist,bins = np.histogram(img_cropped.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(img_cropped.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

6. Klasifikasi Jarak (Euclidean)

```python
# intializing points in
# numpy arrays
point1 = np.array((1, 2, 3))
point2 = np.array((1, 1, 1))
  
# calculating Euclidean distance
# using linalg.norm()
dist = np.linalg.norm(point1 - point2)
  
# printing Euclidean distance
print(dist)
```
